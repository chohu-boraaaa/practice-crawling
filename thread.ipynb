```python
# ÌïÑÏöî ÎùºÏù¥Î∏åÎü¨Î¶¨ import
import threading
```

```python
def sum(low, high):
    total = 0
    for i in range(low, high):
        total += i
    print("Subthread", total)
```

```python
t = threading.Thread(target=sum, args=(1, 100000))
t.start()

print("Main Thread")
```

```python
def run(a, b, ws, wb):
    for i in range(a,b):
        print(i, i+1)
        ws.append(i)
        wb.append(i+1)
```

```python
b=1200
ws = []
wb = []
threads = []
for t in range(1,b,400):
    print(t)
    threads.append(threading.Thread(target=run, args=(t, t+400, ws ,wb)))

for t in threads:
    t.start()
```

```python
import time
import threading
 
def do_something():
    print('Sleeping 1 seconds')
    time.sleep(1)
    print('Done Sleeping...')
 
if __name__ == '__main__':
    start = time.perf_counter()
    threads = []
    for _ in range(10):
        t = threading.Thread(target=do_something)
        t.start()
        threads.append(t)
        
    for thread in threads:
        thread.join()
    finish = time.perf_counter()
 
    print(f'Finished in {round(finish-start, 2)} second(s)')
```
