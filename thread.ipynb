## 1.1. 스레드
- 프로세스 실행 단위
- 프로세스는 최소 하나 이상의 스레드 갖으면 경우에 따라 여러 스레드 가질 수 있음.
- 프로그램 작성 시 멀티 스레드 형태로 구현하면 운영체제에 의해 동시에 스케줄링 될 수 있기 때문에 보통 성능이 더 좋아진다.

## 1.2. Python threading 모듈
- 파이썬 프로그램은 기본적으로 하나의 쓰레드에서 실행되지만, 코드를 병렬로 실행하기 위해서는 별도의 쓰레드를 생성해야 한다.
- threading.Thread() 함수 호출하여 Thread 객체를 얻은 후 Thread 객체의 start() 메서드를 호출하면 된다.

### 1.2.1. thread가 실행할 함수를 작성하고 그 함수 명을 threading.Thread() 함수의 target 아큐먼트에 지정하기
- sum 함수를 쓰레드가 실행하도록 threading.Thread() 함수 파라미터로 target=sum 지정하기
- ❗ 주의 : target=sum()처럼 하면 sum() 함수를 실행하여 리턴 결과를 target에 지정하는 것이므로 잘못된 결과 초래.
- 만약 파라미터 지정해야 하면 args(튜플), kwargs(dict)로 전달하면 된다.

```python
# 필요 라이브러리 import
import threading
```

```python
def sum(low, high):
    total = 0
    for i in range(low, high):
        total += i
    print("Subthread", total)
```

```python
t = threading.Thread(target=sum, args=(1, 100000))
t.start()

print("Main Thread")
```

```python
def run(a, b, ws, wb):
    for i in range(a,b):
        print(i, i+1)
        ws.append(i)
        wb.append(i+1)
```

```python
b=1200
ws = []
wb = []
threads = []
for t in range(1,b,400):
    print(t)
    threads.append(threading.Thread(target=run, args=(t, t+400, ws ,wb)))

for t in threads:
    t.start()
```

### 1.2.2 thread 효과
- 시간 단축 가능!
- 출처 : [https://zephyrus1111.tistory.com/111](https://zephyrus1111.tistory.com/111)

```python
import time
 
def do_something():
    print('Sleeping 1 seconds')
    time.sleep(1)
    print('Done Sleeping...')
 
if __name__ == '__main__':
    start = time.perf_counter()
    for _ in range(10):
        do_something()
    finish = time.perf_counter()
    
    print(f'Finished in {round(finish-start, 2)} second(s)')
```

```python
import time
import threading
 
def do_something():
    print('Sleeping 1 seconds')
    time.sleep(1)
    print('Done Sleeping...')
 
if __name__ == '__main__':
    start = time.perf_counter()
    threads = []
    for _ in range(10):
        t = threading.Thread(target=do_something)
        t.start()
        threads.append(t)
        
    for thread in threads:
        thread.join()
    finish = time.perf_counter()
 
    print(f'Finished in {round(finish-start, 2)} second(s)')
```
